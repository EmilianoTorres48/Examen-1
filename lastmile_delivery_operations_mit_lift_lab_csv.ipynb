```python
data=pd.read_csv("lastmile_delivery_operations_mit_lift_lab_serving_customer.csv")
```

```python
data.info()
```

```python
#Corroboramos valores nulos 
valores_nulos=data.isnull().sum()
valores_nulos
```

PROCEDIMIENTO BFILL

```python
#Reemplazamos valores nulos del data frame con "bfill"
data1 =data.fillna(method="bfill")
data1
```

```python
#Corroboramos valores nulos del data frame
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores nulos del dataframe con "mean"
# Realizamos una copia del data frame
data2=data.copy()
data2=data2.fillna(round(data.mean(),1))
data2
```

```python
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Realizamos diagrama de caja o bigote de cada columna del data frame
fig=plt.figure(figsize=(15, 8))
data2.plot(kind="box", vert=False)
plt.title("valores atipicos del dataframe")
plt.show()  
```

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=data2
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior=y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos outliers del Dataframe
outliers=data2[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
data3=data2[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data3
```

```python
#Eliminamos columnas innecesarios
data4= data3.drop(["_child_record_id","_record_id","_parent_id","_title","_version","_created_at","_updated_at","_created_by","_updated_by","name_of_the_store","entering_the_store","picture_cargo_being_delivered","picture_cargo_being_delivered_captions","picture_cargo_being_delivered_urls","trolley_in_use","leaving_the_store","_geometry","_changeset_id"], axis=1)
data4
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=data4.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del data frame con "mean"
#Realizamos una copia del dataframe
data_clean=data4.copy()
data_clean=data_clean.fillna(round(data4.mean(),1))
data_clean
```

```python
#Corroboramos valores del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

PROCEDIMIENTO "CUANTILES" PARA SUSTUIR OUTLIERS EN DATAFRAME

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data2
percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 -percentile25
Limite_Superior_iqr=percentile75 + 1.5*iqr
Limite_Inferior_iqr=percentile25 - 1.5*iqr
print("Limite superior permitido",Limite_Superior_iqr)
print("Limite inferior permitido",Limite_Inferior_iqr)
```

```python
#Encontramo Outliers del Dataframe
outliers_iqr= data2[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= data2[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
#Eliminamos columnas innecesaris
data4_iqr= data3_iqr.drop(["_child_record_id","_record_id","_parent_id","_title","_version","_created_at","_updated_at","_created_by","_updated_by","name_of_the_store","entering_the_store","picture_cargo_being_delivered","picture_cargo_being_delivered_captions","picture_cargo_being_delivered_urls","trolley_in_use","leaving_the_store","_geometry","_changeset_id"], axis=1)
data4_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data4_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos(nulos)dael dataframe con"mean"
#Realizamos una copia del dataframe
data5_iqr=data4_iqr.copy()
data5_iqr=data5_iqr.fillna(round(data4_iqr.mean(),1))
data5_iqr
```

```python
#Unimos la columna "indice _tiempo" con el dataframe
indice_tiempo=data2["_index"]
Datos_limpios=pd.concat([indice_tiempo, data5_iqr], axis=1)
Datos_limpios
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=Datos_limpios.isnull().sum()
valores_nulos
```

```python
#Convertir dataframe a CSV
Datos_limpios.to_csv("Examen 3.csv")
```
